```python
#import pandas and numpy
import pandas as pd
import numpy as np
```

```python
crime=pd.read_csv("crime.csv")
crime.head()
```

```python
crime.describe() #untuk statdes di phyton
```

```python
np.sum(crime.isnull()) #liat data yang missing value
```

```python
crime['GEO_X']=crime['GEO_X'].fillna(3.162198e+06)
crime['GEO_Y']=crime['GEO_Y'].fillna(1.694853e+06)
crime['GEO_LON']=crime['GEO_LON'].fillna(-104.931834)
crime['GEO_LAT']=crime['GEO_LAT'].fillna(39.720386)
np.sum(crime.isnull())
```

```python
crime['GEO_X']=crime['GEO_X'].fillna((crime['GEO_X'].mean()))
crime['GEO_Y']=crime['GEO_Y'].fillna((crime['GEO_Y'].mean()))
crime['GEO_LON']=crime['GEO_LON'].fillna((crime['GEO_LON'].mean()))
crime['GEO_LAT']=crime['GEO_LAT'].fillna((crime['GEO_LAT'].mean()))
np.sum(crime.isnull())
```

```python
from scipy.stats import mode
#fill missing value with mode for qualitative/categorical data
crime['LAST_OCCURRENCE_DATE']=crime['LAST_OCCURRENCE_DATE'].fillna(crime['LAST_OCCURRENCE_DATE']).dropna().mode()[0]
crime['INCIDENT_ADDRESS']=crime['INCIDENT_ADDRESS'].fillna(crime['INCIDENT_ADDRESS']).dropna().mode()[0]
```

```python
crime.isnull().sum()
```

```python
crime['LAST_OCCURRENCE_DATE'].mode()
crime['INCIDENT_ADDRESS'].mode()
```

```python
##mengatasi outlier
from scipy import stats
z=np.abs(stats.zscore(crime._get_numeric_data()))
print(z)
crime=crime[(z<3).all(axis=1)]
print(crime.shape) #melihat dimensi data
```
