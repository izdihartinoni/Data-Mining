## a. missing value

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("carfuel.csv",sep=";")
data.head()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
data['temp_inside']=data['temp_inside'].fillna(39.671835)
data['temp_outside']=data['temp_outside'].fillna(20.614058)
data['refill liters']=data['refill liters'].fillna(0.334320)
np.sum(data.isnull())
```

```python
from scipy.stats import mode
data['specials']=data['specials'].fillna(data['specials']).dropna().mode()[0]
data.isnull().sum()
```

```python
from scipy import stats
data1=data.drop(['gas_type','refill gas','Unnamed: 12'],1)
data1
```

```python
data1.to_csv('datamissingvalue1.csv')
```

## b. missing value, outlier

```python
# missing value
data1.head()
```

```python
#oulier
from scipy import stats
data1=data.drop(['gas_type','refill gas','Unnamed: 12'],1)
z=np.abs(stats.zscore(data1._get_numeric_data()))
print(z)
dataout=data1[(z<3).all(axis=1)]
print(dataout.shape) 
```

```python
dataout.to_csv('dataoutlier1.csv')
```

## c. missing value, transformasi

```python
# missing value
data1.head()
```

```python
x=data1.iloc[0:388,0:6]
x.tail()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm L2, other are L1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
#minmax
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-5,1))
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#softmax
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```

```python
xsoft.to_csv('datatransformasi1.csv')
```
